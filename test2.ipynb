# 생존분석

생존분석이란?
생존분석은 특정 이벤트(여기서는 퇴사)가 발생할 때까지 걸리는 시간을 분석하는 통계적 방법입니다.
HR 분야에서는 직원 이탈(퇴사)을 예측하고 이해하는 데 매우 유용합니다.

1. 생존함수(Survival Function)
정의: 특정 시점 t까지 이벤트(퇴사)가 발생하지 않을 확률
수식: S(t) = P(T > t)
해석: t시점에서의 값이 0.8이라면, "입사 후 t개월까지 80%의 직원이 남아있다"는 의미


        2022-01-01~2023-12-31에 입사한 사람들을 분석하는데
        특정 시점(x축 2년)에서의 생존률(30%)을 계산할 때

        포함되는 경우 (생존율 30%에 포함됨)
        1. 2022년에 입사하여 2024-12-31 현재까지 재직 중인 사람
        2. 2022년에 입사하여 현재는 퇴사했지만 2년 넘게 근무한 후 퇴사한 사람

        제외되는 경우 (생존율 30% 계산에 포함되지 않음)
        1. 2023년에 입사하여 2024-12-31 현재까지 근무 중이지만, 아직 입사한 지 2년이 안 된 사람
        2. 입사 후 2년이 되기 전에 퇴사한 사람

2. 중도절단(Censoring)

정의: 관찰 기간이 끝날 때까지 이벤트(퇴사)가 발생하지 않은 경우
코드에서의 처리: 퇴사일이 없는 직원은 2024년 말까지 재직한 것으로 가정하고 중도절단 데이터로 처리

3. 카플란-마이어 추정법(Kaplan-Meier Estimator)

정의: 비모수적 방법으로 생존함수를 추정하는 기법
구현: calculate_survival_function() 함수에서 구현됨
계산 방식: 각 시점마다 생존해 있는 직원 수 / 전체 직원 수

생존곡선 해석 방법

1. 기본 해석 요소
생존확률(Y축)

의미: 특정 시점까지 직원이 회사에 남아있을 확률
시작점: 일반적으로 1.0(100%)에서 시작
해석 예시:
12개월 시점에서 0.85라면 → "입사 후 1년 시점에 85%의 직원이 여전히 재직 중"
36개월 시점에서 0.42라면 → "입사 후 3년까지 42%의 직원만이 회사에 남아있음"

이미 퇴사한 사람들: 생존분석에서 "사건 발생(event)"으로 처리되며, 이들의 실제 재직기간이 생존함수 계산에 반영
아직 재직 중인 사람들: "중도절단(censored)" 데이터로 처리되며, 이들은 현재까지 관찰된 재직기간까지만 생존함수 계산에 기여
따라서 '3년 지점에서 생존율 20%'라는 수치에는:
1) 실제로 3년 이상 재직한 후 퇴사한 직원들
2) 현재 3년 이상 재직 중인 직원들
3) 아직 3년이 안 됐지만 현재까지 퇴사하지 않은 직원들

시간축(X축)

재직 기간을 나타내며, 코드에서는 월 단위로 계산하고 표시는 년 단위로 변환
특정 시점에서 수직선을 그으면 해당 시점까지의 직원 유지율을 확인 가능

2. 곡선 형태별 해석
급격한 하락 구간

의미: 퇴사가 집중적으로 발생하는 '고위험 기간'
해석:
입사 후 3-6개월 사이에 급격한 하락 → 온보딩 프로그램이나 초기 적응에 문제 있음
12-18개월 사이 급격한 하락 → 1년 차 평가/승진 시점에서의 이탈 발생
24-30개월 사이 급격한 하락 → 경력 성장 한계에 봉착한 중간 경력자들의 이탈

완만한 하락 구간
의미: 안정적인 유지율을 보이는 기간
해석:
36개월 이후 완만한 하락 → 장기 재직자들은 상대적으로 안정적인 유지율 보임
전체적으로 완만한 곡선 → 전반적인 조직 안정성이 높음

평탄 구간(Plateau)
의미: 퇴사가 거의 없는 안정기
해석:
특정 기간 동안 평탄한 구간 → 해당 기간 동안 퇴사가 거의 없었음
장기간 지속되는 평탄 구간 → 일정 기간 재직 후 매우 높은 직원 충성도 형성

계단식 하락
의미: 특정 시점에 집중적인 퇴사 발생
해석:
연말/연초에 계단식 하락 → 보너스 수령 후 퇴사하는 패턴
분기별 계단식 하락 → 분기 평가 후 퇴사 결정하는 패턴

3. 고급 분석 기법
중앙 생존 시간(Median Survival Time)

정의: 생존확률이 0.5(50%)가 되는 시점
계산 방법: 생존곡선에서 Y축 0.5 지점과 만나는 X축 값
해석:
중앙 생존 시간이 18개월이라면 → "직원들의 절반이 18개월 내에 퇴사함"
직군 A의 중앙 생존 시간: 24개월, 직군 B: 36개월 → "직군 A가 B보다 이직률이 높음"

특정 시점 생존율 비교
방법: 동일한 시간 지점(예: 1년, 2년)에서 여러 곡선의 생존확률 비교
해석 예시:
2년 시점에서 개발직군 생존율 65%, 영업직군 45% → "개발직군이 영업직군보다 2년 유지율이 20%p 높음"
모든 직군이 1년 시점에서 급격히 하락 → "1년 차는 전사적으로 중요한 이탈 시점"

생존곡선 간 격차 분석
의미: 직군/부서/성별 등 그룹 간 생존곡선 차이
격차 확대 구간: 두 곡선 간 차이가 커지는 구간 → 해당 기간에 그룹 간 차별적 요인 발생
격차 축소 구간: 두 곡선이 수렴하는 구간 → 장기적으로는 그룹 간 차이가 줄어듦

4. 실무적 해석 사례
입사 첫해 생존곡선 해석
관찰: 입사 후 3-6개월 사이 생존율이 90%에서 70%로 급락
해석: 온보딩 프로그램이 3개월까지만 효과적이며, 이후 지원 부족
대응 방안: 6개월까지 온보딩 프로그램 연장, 3-6개월 차 직원 대상 멘토링 강화

직군별 생존곡선 비교 해석
관찰: IT 직군은 2년 차에 급격한 하락, 영업직군은 1년 차에 급격한 하락
해석:
IT: 기술 성장 한계 도달 후 이직 시도
영업: 첫 성과평가 후 성과 부진자 이탈
대응 방안:
IT: 2년 차 기술 역량 개발 프로그램 도입
영업: 첫해 성과 관리 및 코칭 강화

연령대별 생존곡선 해석
관찰: 20대는 가파른 하락, 30-40대는 완만한 하락, 50대 이상은 매우 안정적
해석: 젊은 세대의 경력 탐색 성향과 중장년층의 안정 지향 성향 반영
대응 방안: 젊은 세대에게는 다양한 경력 경로 제공, 중장년층에게는 안정성 강조

6. 시계열적 변화 해석
연도별 생존곡선 비교
관찰: 2022년 코호트보다 2023년 코호트의 생존곡선이 더 가파르게 하락
해석:
노동시장 변화(구직 기회 증가)
채용 정책 변화로 인한 적합성 저하
코로나 이후 원격근무 정책 변화 영향
분석 방법: 각 연도별 코호트의 동일 시점(예: 6개월, 1년) 생존율 추세 비교

정책 변화 전후 비교
관찰: 복지 정책 개선 후 입사한 코호트의 1년 생존율이 8%p 향상
해석: 정책 변화가 직원 유지에 긍정적 영향을 미침
효과 측정: 정책 변화 시점 이전/이후 코호트의 생존곡선 기울기 변화 분석

7. 예측적 해석
미래 인력 예측
현재의 생존곡선을 기반으로 향후 퇴사율 예측
신규 채용 100명 중 2년 후 예상 잔존 인원 = 100 × (2년 시점 생존율)

```python
import os
import pandas as pd
import numpy as np
from datetime import datetime

import dash
import dash_bootstrap_components as dbc
from dash import dcc, html, Input, Output, State, callback, ctx
import dash_bootstrap_components as dbc
import plotly.graph_objects as go
import plotly.express as px
```

```python

# 애플리케이션 초기화
app = dash.Dash(__name__, 
                external_stylesheets=[dbc.themes.BOOTSTRAP],
                suppress_callback_exceptions=True,
                meta_tags=[{'name': 'viewport', 'content': 'width=device-width, initial-scale=1'}])
server = app.server
app.title = '퇴사자 생존분석 대시보드'

# 전역 변수 및 상수 정의
DATA_PATH = '퇴사자분석.xlsx'  # 데이터 파일 경로
MAX_SELECTIONS = 2  # 최대 선택 가능한 분석 관점 수

# 데이터 로드 및 전처리 함수
def load_and_preprocess_data(file_path):
    """
    엑셀 파일을 로드하고 분석에 필요한 형태로 전처리합니다.
    
    Parameters:
    file_path (str): 엑셀 파일 경로
    
    Returns:
    tuple: (전처리된 데이터프레임, 최신 지급년월)
    """
    try:
        # 엑셀 파일 로드
        data = pd.read_excel(file_path, engine='openpyxl')
        
        # 날짜 열 변환
        data['입사일'] = pd.to_datetime(data['입사일'])
        data['퇴사일'] = pd.to_datetime(data['퇴사일'])
        
        # 최신 지급년월 확인
        max_yearmonth = data['지급년월'].max()
        latest_data = data[data['지급년월'] == max_yearmonth]
        
        # 재직 중인 직원 목록 식별
        current_employees = latest_data[latest_data['퇴사일'].isna()]['성명'].unique()
        
        # 각 직원의 마지막 기록만 추출
        last_records = data.sort_values('지급년월').groupby('성명').last().reset_index()
        
        # 재직기간 계산 함수
        def calculate_tenure(row):
            """직원의 재직기간(월)을 계산하는 함수"""
            end_date = row['퇴사일'] if pd.notna(row['퇴사일']) else pd.Timestamp('2024-12-31')
            if pd.notna(row['입사일']):
                # 평균 월 길이(30.44일)로 나누어 월 단위 계산
                tenure = (end_date - row['입사일']).days / 30.44
                return max(0, tenure)  # 음수 값 방지
            return np.nan
        
        # 재직기간 계산 및 재직중 여부 표시
        last_records['재직기간_월'] = last_records.apply(calculate_tenure, axis=1)
        last_records['재직중'] = last_records['성명'].isin(current_employees)
        
        # 결측치 처리: 직군
        if '직군' in last_records.columns:
            last_records['직군'] = last_records['직군'].fillna('미정')
        
        # 결측치 처리: 사업장명
        if '사업장명' in last_records.columns:
            last_records['사업장명'] = last_records['사업장명'].fillna('미정')
        
        # 결측치 처리: 성별
        if '성별2' in last_records.columns:
            last_records['성별'] = last_records['성별2'].fillna('미정')
        elif '성별' in last_records.columns:
            last_records['성별'] = last_records['성별'].fillna('미정')
            
        # 결측치 처리: 직무성격
        if '직무성격(대)' in last_records.columns:
            last_records['직무성격(대)'] = last_records['직무성격(대)'].fillna('미정')
        if '직무성격(중)' in last_records.columns:
            last_records['직무성격(중)'] = last_records['직무성격(중)'].fillna('미정')
        
        # 연령대 구분 함수
        def get_age_group(age):
            """연령을 기준으로 연령대를 구분하는 함수"""
            if pd.isna(age):
                return '연령 미상'
            elif age < 30:
                return '20대'
            elif age < 40:
                return '30대'
            elif age < 50:
                return '40대'
            else:
                return '50대 이상'
        
        # '연령대' 열 추가
        last_records['연령대'] = last_records['연령'].apply(get_age_group)
        
        return last_records, max_yearmonth
    
    except Exception as e:
        print(f"데이터 로드 중 오류 발생: {e}")
        # 오류 발생 시 빈 데이터프레임 반환
        return pd.DataFrame(), None

# 생존함수 계산 함수
def calculate_survival_function(durations, events):
    """
    생존함수(생존확률)를 계산합니다.
    
    Parameters:
    durations (array): 재직기간(월) 배열
    events (array): 퇴사 이벤트 발생 여부 배열 (True=퇴사, False=재직중)
    
    Returns:
    tuple: (unique_durations, survival_probabilities)
    """
    unique_durations = np.sort(np.unique(durations))
    survival_prob = []
    n_total = len(durations)
    
    for t in unique_durations:
        # t시점 이상 재직 중인 직원 수 계산
        n_survived = np.sum(durations >= t)
        # 생존확률 계산 (전체 인원 대비 t시점 이상 재직 중인 비율)
        survival_prob.append(n_survived / n_total if n_total > 0 else 1)
    
    return unique_durations, np.array(survival_prob)

# 생존곡선 생성 함수
def create_survival_curve(data, start_date, end_date, analysis_dimensions, selected_values):
    """
    선택된 분석 기준과 조건으로 생존곡선을 생성합니다.
    
    Parameters:
    data (DataFrame): 전처리된 데이터
    start_date (datetime): 시작 날짜
    end_date (datetime): 종료 날짜
    analysis_dimensions (list): 선택된 분석 관점 목록 (예: ['직군', '성별'])
    selected_values (dict): 각 분석 관점별 선택된 값들 (예: {'직군': ['개발', '관리'], '성별': ['남', '여']})
    
    Returns:
    go.Figure: 생성된 생존곡선 그래프
    """
    # 날짜 범위로 데이터 필터링
    filtered_data = data[
        (data['입사일'] >= start_date) &
        (data['입사일'] <= end_date)
    ]
    
    # 각 차원별 필터링
    for dim in analysis_dimensions:
        if dim in selected_values and selected_values[dim]:
            filtered_data = filtered_data[filtered_data[dim].isin(selected_values[dim])]
    
    # 필터 결과가 없으면 빈 Figure 반환
    if len(filtered_data) == 0:
        fig = go.Figure()
        fig.update_layout(
            title='데이터 없음',
            annotations=[{
                'text': '선택한 조건에 해당하는 데이터가 없습니다',
                'showarrow': False,
                'font': {'size': 20},
                'xref': 'paper',
                'yref': 'paper',
                'x': 0.5,
                'y': 0.5
            }]
        )
        return fig
    
    fig = go.Figure()
    
    # 1차원 분석인 경우 (관점이 1개만 선택된 경우)
    if len(analysis_dimensions) == 1:
        dimension = analysis_dimensions[0]
        unique_values = sorted(filtered_data[dimension].unique())
        
        # 색상 팔레트 - 최대 20색 지원
        colorscale = px.colors.qualitative.Plotly
        
        # 각 값별 생존곡선 생성
        for idx, value in enumerate(unique_values):
            value_data = filtered_data[filtered_data[dimension] == value]
            
            # 생존함수 계산
            durations = value_data['재직기간_월'].values
            events = ~value_data['재직중'].values  # 재직중이면 이벤트(퇴사) 미발생
            timeline, survival = calculate_survival_function(durations, events)
            
            # 차트에 라인 추가
            color = colorscale[idx % len(colorscale)]
            fig.add_trace(go.Scatter(
                x=timeline,
                y=survival,
                name=f"{value} (n={len(value_data)})",
                mode='lines',
                line=dict(width=2, color=color)
            ))
    
    # 2차원 분석인 경우 (관점이 2개 선택된 경우)
    elif len(analysis_dimensions) == 2:
        dim1, dim2 = analysis_dimensions
        
        # 두 차원의 모든 조합 찾기
        dim_combos = filtered_data.groupby([dim1, dim2]).size().reset_index().rename(columns={0: 'count'})
        
        # 색상 팔레트
        colorscale = px.colors.qualitative.Plotly
        
        # 각 조합별 생존곡선 생성
        for idx, combo in dim_combos.iterrows():
            value1 = combo[dim1]
            value2 = combo[dim2]
            count = combo['count']
            
            # 해당 조합의 데이터 추출
            combo_data = filtered_data[(filtered_data[dim1] == value1) & (filtered_data[dim2] == value2)]
            
            # 생존함수 계산
            durations = combo_data['재직기간_월'].values
            events = ~combo_data['재직중'].values
            timeline, survival = calculate_survival_function(durations, events)
            
            # 차트에 라인 추가
            color = colorscale[idx % len(colorscale)]
            fig.add_trace(go.Scatter(
                x=timeline,
                y=survival,
                name=f"{value1}-{value2} (n={count})",
                mode='lines',
                line=dict(width=2, color=color)
            ))
    
    # 모든 선택 데이터에 대한 전체 생존곡선 (둘 다 아닌 경우)
    else:
        durations = filtered_data['재직기간_월'].values
        events = ~filtered_data['재직중'].values
        timeline, survival = calculate_survival_function(durations, events)
        fig.add_trace(go.Scatter(
            x=timeline,
            y=survival,
            name=f'전체 (n={len(filtered_data)})',
            mode='lines',
            line=dict(width=3, color='black')
        ))
    
    # x축 설정 (최대 재직기간 기준으로 1년 단위 올림)
    max_duration = filtered_data['재직기간_월'].max() if len(filtered_data) > 0 else 12
    x_max = np.ceil(max_duration / 12) * 12
    xticks = np.arange(0, x_max + 12, 12)
    
    # 그래프 레이아웃 설정
    title_text = ' & '.join(analysis_dimensions) + ' 생존분석' if analysis_dimensions else '전체 생존분석'
    fig.update_layout(
        title=title_text,
        xaxis=dict(
            title='재직기간',
            range=[0, x_max],
            tickvals=xticks,
            ticktext=[f'{int(x/12)}년' for x in xticks]
        ),
        yaxis=dict(
            title='생존확률',
            range=[0, 1],
            tickformat='.0%'
        ),
        hovermode='x unified',
        showlegend=True,
        legend=dict(
            orientation="v",
            yanchor="top",
            y=0.99,
            xanchor="left",
            x=1.05,
        ),
        template='plotly_white',
        margin=dict(r=150)  # 범례 공간 확보
    )
    
    # 평균 재직기간 표시 (데이터가 있는 경우만)
    if len(filtered_data) > 0:
        avg_tenure_months = filtered_data['재직기간_월'].mean()
        avg_tenure_years = avg_tenure_months / 12
        fig.add_vline(
            x=avg_tenure_months, 
            line=dict(color='red', dash='dash'),
            annotation_text=f"평균재직기간: {avg_tenure_years:.1f}년",
            annotation_position="top right"
        )
    
    return fig

# 통계 HTML 생성 함수
def create_statistics_html(filtered_data, analysis_dimensions, selected_values, max_yearmonth):
    """
    선택된 분석 기준과 조건으로 통계 HTML을 생성합니다.
    
    Parameters:
    filtered_data (DataFrame): 필터링된 데이터
    analysis_dimensions (list): 선택된 분석 관점 목록
    selected_values (dict): 각 분석 관점별 선택된 값들
    max_yearmonth: 최신 지급년월
    
    Returns:
    str: HTML 형식의 통계 정보
    """
    if len(filtered_data) == 0:
        return "<h3>선택된 조건에 맞는 데이터가 없습니다.</h3>"
    
    html = f"<h3>선택된 기간 및 조건의 기본 통계 (기준 시점: {max_yearmonth})</h3>"
    html += f"<p>전체 직원 수: {len(filtered_data)}</p>"
    html += f"<p>현재 재직자 수: {filtered_data['재직중'].sum()}</p>"
    html += f"<p>퇴사자 수: {(~filtered_data['재직중']).sum()}</p>"
    html += f"<p>평균 재직기간: {filtered_data['재직기간_월'].mean()/12:.1f}년</p>"
    html += f"<p>중앙 재직기간: {filtered_data['재직기간_월'].median()/12:.1f}년</p>"
    
    # 2차원 분석인 경우 (관점이 2개 선택된 경우)
    if len(analysis_dimensions) == 2:
        dim1, dim2 = analysis_dimensions
        html += f"<h3>{dim1}-{dim2} 조합별 통계</h3>"
        
        # 두 차원의 조합별 통계 계산
        combo_stats = filtered_data.groupby([dim1, dim2]).agg({
            '재직중': ['count', 'sum'],
            '재직기간_월': ['mean', 'median']
        }).round(1)
        combo_stats.columns = ['총 인원', '재직자 수', '평균 재직기간(월)', '중앙 재직기간(월)']
        combo_stats['퇴사자 수'] = combo_stats['총 인원'] - combo_stats['재직자 수']
        combo_stats['퇴사율'] = ((combo_stats['퇴사자 수']) / combo_stats['총 인원'] * 100).round(1)
        combo_stats['평균 재직기간(년)'] = (combo_stats['평균 재직기간(월)'] / 12).round(1)
        combo_stats['중앙 재직기간(년)'] = (combo_stats['중앙 재직기간(월)'] / 12).round(1)
        
        # 테이블 형식으로 변환
        html += combo_stats.to_html(classes="table table-striped table-hover")
    
    # 각 차원별 통계 추가
    for dim in analysis_dimensions:
        html += f"<h3>{dim}별 통계</h3>"
        
        dim_stats = filtered_data.groupby(dim).agg({
            '재직중': ['count', 'sum'],
            '재직기간_월': ['mean', 'median']
        }).round(1)
        dim_stats.columns = ['총 인원', '재직자 수', '평균 재직기간(월)', '중앙 재직기간(월)']
        dim_stats['퇴사자 수'] = dim_stats['총 인원'] - dim_stats['재직자 수']
        dim_stats['퇴사율'] = ((dim_stats['퇴사자 수']) / dim_stats['총 인원'] * 100).round(1)
        dim_stats['평균 재직기간(년)'] = (dim_stats['평균 재직기간(월)'] / 12).round(1)
        dim_stats['중앙 재직기간(년)'] = (dim_stats['중앙 재직기간(월)'] / 12).round(1)
        
        # 특별한 순서가 필요한 경우 처리 (예: 연령대)
        if dim == '연령대':
            age_group_order = ['20대', '30대', '40대', '50대 이상', '연령 미상']
            # 존재하는 그룹만 정렬
            ordered_idx = [idx for idx in age_group_order if idx in dim_stats.index]
            dim_stats = dim_stats.reindex(ordered_idx)
        
        html += dim_stats.to_html(classes="table table-striped table-hover")
    
    return html

# 데이터 로드 (앱 시작 시 한 번만 실행)
try:
    print("데이터 로드 중...")
    df, max_yearmonth = load_and_preprocess_data(DATA_PATH)
    print(f"데이터 로드 완료: {len(df)}개 레코드")
    
    # 분석 관점 정의 (실제 데이터에서 사용 가능한 컬럼만 포함)
    available_dimensions = [
        {'label': '직군', 'value': '직군'},
        {'label': '사업장명', 'value': '사업장명'},
        {'label': '성별', 'value': '성별'},
        {'label': '연령대', 'value': '연령대'},
    ]
    
    # 직무성격 컬럼이 있는 경우 추가
    if '직무성격(대)' in df.columns:
        available_dimensions.append({'label': '직무성격(대)', 'value': '직무성격(대)'})
    if '직무성격(중)' in df.columns:
        available_dimensions.append({'label': '직무성격(중)', 'value': '직무성격(중)'})
        
    # 각 분석 관점별 가능한 값 미리 계산
    dimension_values = {}
    for dim in [d['value'] for d in available_dimensions]:
        if dim in df.columns:
            dimension_values[dim] = sorted(df[dim].unique())
    
    # 날짜 범위 설정
    min_date = df['입사일'].min() if not df.empty else datetime(2022, 1, 1)
    max_date = df['입사일'].max() if not df.empty else datetime(2024, 12, 31)
    
except Exception as e:
    print(f"초기 데이터 로드 오류: {e}")
    df = pd.DataFrame()
    max_yearmonth = "데이터 없음"
    available_dimensions = []
    dimension_values = {}
    min_date = datetime(2022, 1, 1)
    max_date = datetime(2024, 12, 31)

# 레이아웃 정의
# 모든 필터를 미리 넣어두고, 선택된 관점이 아니면 숨기는 방식
all_filter_components = []
for dim in available_dimensions:
    dim_key = dim['value']
    all_filter_components.append(
        html.Div([
            html.Label(f"{dim_key} 선택:", className="mt-2"),
            dcc.Dropdown(
                id=f'filter-{dim_key}',
                options=[{'label': val, 'value': val} for val in dimension_values.get(dim_key, [])],
                value=dimension_values.get(dim_key, []),  # 기본적으로 모두 선택
                multi=True
            ),
            html.Div([
                dbc.Button(
                    "모두 선택", 
                    id=f'select-all-{dim_key}', 
                    color="info", 
                    size="sm",
                    className="me-2 mt-1"
                ),
                dbc.Button(
                    "모두 해제", 
                    id=f'deselect-all-{dim_key}', 
                    color="warning", 
                    size="sm",
                    className="me-2 mt-1"
                )
            ])
        ],
        id=f'div-filter-{dim_key}',  # 이 div 전체를 보여줄지 말지 결정
        className="mb-3")
    )

app.layout = dbc.Container([
    dbc.Row([
        dbc.Col([
            html.H1("퇴사자 생존분석 대시보드", className="text-center mb-4")
        ], width=12)
    ]),
    
    dbc.Row([
        dbc.Col([
            html.Div([
                html.H4("분석 설정", className="mb-3"),
                html.Label("분석 관점 선택 (최대 2개):", className="fw-bold"),
                dcc.Checklist(
                    id='analysis-dimensions',
                    options=available_dimensions,
                    value=[],
                    labelStyle={'display': 'inline-block', 'margin-right': '10px'},
                    className="mb-3"
                ),
                
                # 모든 필터 컴포넌트는 여기에 미리 배치
                html.Div(
                    id='dimension-filters-container',
                    children=all_filter_components
                ),
                
                html.Hr(),
                
                html.Label("입사일 기간 설정:", className="fw-bold"),
                dbc.Row([
                    dbc.Col([
                        html.Label("시작일:"),
                        dcc.DatePickerSingle(
                            id='date-picker-start',
                            min_date_allowed=min_date,
                            max_date_allowed=max_date,
                            date=min_date,
                            display_format='YYYY-MM-DD'
                        )
                    ], width=6),
                    dbc.Col([
                        html.Label("종료일:"),
                        dcc.DatePickerSingle(
                            id='date-picker-end',
                            min_date_allowed=min_date,
                            max_date_allowed=max_date,
                            date=max_date,
                            display_format='YYYY-MM-DD'
                        )
                    ], width=6)
                ], className="mb-3"),
                
                dbc.Button(
                    "분석 실행", 
                    id="run-analysis-button", 
                    color="primary",
                    className="me-2 mt-2"
                ),
                
                html.Div(id='data-info', children=[
                    html.P(f"데이터 기준 시점: {max_yearmonth}", className="text-muted mt-2")
                ])
            ], className="p-3 border rounded")
        ], width=12, lg=3),
        
        dbc.Col([
            dbc.Tabs([
                dbc.Tab([
                    html.Div(id='stats-container', className="pt-3")
                ], label="통계 정보"),
                dbc.Tab([
                    dcc.Loading(
                        id="loading-graph",
                        type="default",
                        children=dcc.Graph(id='survival-graph', style={'height': '80vh'})
                    )
                ], label="생존 분석 차트")
            ])
        ], width=12, lg=9)
    ]),
    
    dbc.Row([
        dbc.Col([
            html.Hr(),
            html.P(
                "대시보드 구현: 퇴사자 생존분석 도구 © 2024", 
                className="text-center text-muted"
            )
        ], width=12)
    ])
], fluid=True)

# 콜백: 분석 관점이 바뀔 때, 각 필터 div를 보이거나 숨긴다.
@callback(
    [Output(f'div-filter-{dim["value"]}', 'style') for dim in available_dimensions],
    Input('analysis-dimensions', 'value')
)
def toggle_filter_visibility(selected_dimensions):
    """선택된 관점만 필터를 보이게 하고, 나머지는 숨긴다."""
    results = []
    for dim in available_dimensions:
        if dim['value'] in selected_dimensions:
            results.append({'display': 'block'})
        else:
            results.append({'display': 'none'})
    return results

# 선택된 분석 관점이 MAX_SELECTIONS를 초과하지 않도록 제한
@callback(
    Output('analysis-dimensions', 'value', allow_duplicate=True),
    Input('analysis-dimensions', 'value'),
    prevent_initial_call=True
)
def limit_dimensions(selected_dimensions):
    """선택된 분석 관점의 수를 제한합니다."""
    if len(selected_dimensions) > MAX_SELECTIONS:
        return selected_dimensions[:MAX_SELECTIONS]
    return selected_dimensions

# 각 차원별 '모두 선택' 버튼 콜백
for dim in [d['value'] for d in available_dimensions]:
    @callback(
        Output(f'filter-{dim}', 'value'),
        Input(f'select-all-{dim}', 'n_clicks'),
        Input(f'deselect-all-{dim}', 'n_clicks'),
        State(f'filter-{dim}', 'options'),
        prevent_initial_call=True
    )
    def update_dropdown_selection(select_clicks, deselect_clicks, options, dim=dim):
        """모두 선택/해제 버튼 핸들러"""
        trigger = ctx.triggered_id
        if trigger == f'select-all-{dim}':
            return [option['value'] for option in options]
        elif trigger == f'deselect-all-{dim}':
            return []
        return dash.no_update

# 메인 분석 실행 콜백
@callback(
    [Output('survival-graph', 'figure'),
     Output('stats-container', 'children')],
    Input('run-analysis-button', 'n_clicks'),
    [State('analysis-dimensions', 'value'),
     State('date-picker-start', 'date'),
     State('date-picker-end', 'date')] +
    [State(f'filter-{dim["value"]}', 'value') for dim in available_dimensions],
    prevent_initial_call=True
)
def run_analysis(n_clicks, analysis_dimensions, start_date_str, end_date_str, *args):
    """
    분석 실행 버튼 클릭 시 생존분석 차트와 통계 정보를 생성합니다.
    """
    if not n_clicks:
        return dash.no_update, dash.no_update
    
    # 날짜 변환
    start_date = pd.to_datetime(start_date_str)
    end_date = pd.to_datetime(end_date_str)
    
    # 각 분석 관점별 선택된 값들 수집
    dimension_filters = {}
    for i, dim in enumerate(available_dimensions):
        dim_name = dim['value']
        # args 순서대로 매칭
        dimension_filters[dim_name] = args[i]
    
    # 날짜 범위로 데이터 필터링
    filtered_data = df[
        (df['입사일'] >= start_date) &
        (df['입사일'] <= end_date)
    ]
    
    # 생존곡선 그래프 생성
    fig = create_survival_curve(
        filtered_data, 
        start_date, 
        end_date, 
        analysis_dimensions, 
        dimension_filters
    )
    
    # 통계 HTML 생성
    stats_html = create_statistics_html(
        filtered_data, 
        analysis_dimensions, 
        dimension_filters, 
        max_yearmonth
    )
    
    return fig, dcc.Markdown(stats_html, dangerously_allow_html=True)

# 앱 실행
if __name__ == '__main__':
    app.run_server(debug=False, host='0.0.0.0')
```
